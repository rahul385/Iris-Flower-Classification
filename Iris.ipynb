# Iris Flower dataset

## Objective: Classify a new flower as belonging to one of the 3 classes: Iris-setosa, Iris-versicolor, Iris-virginica


```python
# to handle dataset
import pandas as pd
import numpy as np

# to profile the data
from pandas_profiling import ProfileReport

# to visualize data
import matplotlib.pyplot as plt
import seaborn as sns

#to encode labels
from sklearn.preprocessing import LabelEncoder

# to split dataset
from sklearn.model_selection import train_test_split

# feature scaling
from sklearn.preprocessing import StandardScaler

# to train the model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier

# to evaluate the models
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import accuracy_score
```

```python
#Load and view data
df=pd.read_csv('iris.csv')
df.head()
```

### Label Encoding

```python
### label encode the categorical values and convert them to numbers 
le = LabelEncoder()
le.fit(df['Species'].astype(str))
df['Species'] = le.transform(df['Species'].astype(str))
```

```python
df['Species'].unique()
```

### Separate train and test set

```python
# Separate data into train and test set
X_train, X_test, y_train, y_test = train_test_split(df, df.Species, test_size=0.3,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### Outliers

```python
# Outliers
# let's make boxplots to visualise outliers in the continuous variables 
# and histograms to get an idea of the distribution
continuous=('SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm')

for var in continuous:
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    fig = X_train.boxplot(column=var)
    fig.set_title('')
    fig.set_ylabel(var)
    
    plt.subplot(1, 2, 2)
    fig = X_train[var].hist(bins=20)
    fig.set_ylabel('Species')
    fig.set_xlabel(var)

    plt.show()
```

```python
training_vars = [var for var in X_train.columns if var not in ['Species','Id']]
training_vars
```

### Model Training and Evaluation

```python
# Train Model Decision Tree
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train[training_vars], y_train)

#Evaluate Decision Tree
print('Cross Validation score Decision Tree')
print(cross_val_score(tree_model, X_train[training_vars], y_train, cv=3, scoring="accuracy"))

print('Confusion Matrix')
y_train_pred = cross_val_predict(tree_model, X_train[training_vars], y_train, cv=3)
print(confusion_matrix(y_train, y_train_pred))

print('Precision/Recall score')
print(precision_score(y_train, y_train_pred,average=None),recall_score(y_train, y_train_pred,average=None))

print('Accuracy score training set')
print(metrics.accuracy_score(y_train_pred, y_train))

print('Accuracy score test set')
prediction=tree_model.predict(X_test[training_vars]) 
print(metrics.accuracy_score(prediction,y_test))
```

###### Logistic Regression: Training and Evaluation

```python
log_model = LogisticRegression()
log_model.fit(X_train[training_vars], y_train)

#Evaluate XGBoost
print('Cross Validation score')
print(cross_val_score(log_model, X_train[training_vars], y_train, cv=3, scoring="accuracy"))

print('Confusion Matrix')
y_train_pred = cross_val_predict(log_model, X_train[training_vars], y_train, cv=3)
print(confusion_matrix(y_train, y_train_pred))

print('Precision/Recall score')
print(precision_score(y_train, y_train_pred,average=None),recall_score(y_train, y_train_pred,average=None))

print('Accuracy score training set')
print(metrics.accuracy_score(y_train_pred, y_train))

print('Accuracy score test set')

prediction=log_model.predict(X_test[training_vars]) 
print(metrics.accuracy_score(prediction,y_test))
```

###### XGBoost: Training & Evaluation

```python
# Train Model XGBoost
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train[training_vars], y_train)

#Evaluate XGBoost
print('Cross Validation score')
print(cross_val_score(xgb_model, X_train[training_vars], y_train, cv=3, scoring="accuracy"))

print('Confusion Matrix')
y_train_pred = cross_val_predict(xgb_model, X_train[training_vars], y_train, cv=3)
print(confusion_matrix(y_train, y_train_pred))

print('Precision/Recall score')
print(precision_score(y_train, y_train_pred,average=None),recall_score(y_train, y_train_pred,average=None))

print('Accuracy score training set')
print(metrics.accuracy_score(y_train_pred, y_train))

print('Accuracy score test set')
prediction=xgb_model.predict(X_test[training_vars]) 
print(metrics.accuracy_score(prediction,y_test))
```

###### AdaBoost: Training & Evaluation

```python
ada_model = AdaBoostClassifier()
ada_model.fit(X_train[training_vars], y_train)

print('Cross Validation score')
print(cross_val_score(ada_model, X_train[training_vars], y_train, cv=3, scoring="accuracy"))

print('Confusion Matrix')
y_train_pred = cross_val_predict(ada_model, X_train[training_vars], y_train, cv=3)
print(confusion_matrix(y_train, y_train_pred))

print('Precision/Recall score')
print(precision_score(y_train, y_train_pred,average=None),recall_score(y_train, y_train_pred,average=None))

print('Accuracy score training set')
print(metrics.accuracy_score(y_train_pred, y_train))

print('Accuracy score test set')
prediction=ada_model.predict(X_test[training_vars]) 
print(metrics.accuracy_score(prediction,y_test))
```

###### Random Forest: Training & Evaluation

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_train[training_vars], y_train)

print('Cross Validation score')
print(cross_val_score(rf_model, X_train[training_vars], y_train, cv=3, scoring="accuracy"))

print('Confusion Matrix')
y_train_pred = cross_val_predict(rf_model, X_train[training_vars], y_train, cv=3)
print(confusion_matrix(y_train, y_train_pred))

print('Precision/Recall score')
print(precision_score(y_train, y_train_pred,average=None),recall_score(y_train, y_train_pred,average=None))

print('Accuracy score training set')
print(metrics.accuracy_score(y_train_pred, y_train))

print('Accuracy score test set')
prediction=rf_model.predict(X_test[training_vars]) 
print(metrics.accuracy_score(prediction,y_test))
```

### Ensemble Method: Soft Voting Classifier

```python
voting_clf = VotingClassifier(
estimators=[('lr', rf_model), ('rf', xgb_model), ('svc', ada_model),('log', log_model)],
voting='soft'
)
voting_clf.fit(X_train[training_vars], y_train)
```

```python
for clf in (xgb_model,rf_model,ada_model,log_model, voting_clf):
    clf.fit(X_train[training_vars], y_train)
    y_pred = clf.predict(X_test[training_vars])
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

### Optimal Model

###### As can be seen from the results, XGBoost & Random Forest give the highest accuracy.
